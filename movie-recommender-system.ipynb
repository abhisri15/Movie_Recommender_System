# 2. Data Preprocessing

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.iloc[0].genres
```

```python
def fetch_tags(obj):
    List = []
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List
```

```python
df['genres'] = movies['genres'].apply(fetch_tags)
```

```python
df.head()
```

```python
df['keywords'] = df['keywords'].apply(fetch_tags)
```

```python
df.head()
```

```python
def fetch_cast(obj):
    List = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            List.append(i['name'])
            counter += 1
        else:
            break
    return List
```

```python
df['cast'] = df['cast'].apply(fetch_cast)
```

```python
df.head()
```

```python
def fetch_director(obj):
    List = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            List.append(i["name"])
            break
    return List
```

```python
df['crew'] = df['crew'].apply(fetch_director)
```

```python
df.head()
```

```python
df['overview'] = df['overview'].apply(lambda x: x.split())
```

```python
df.head()
```

```python
df['genres'] = df['genres'].apply(lambda x: [i.replace(" ","") for i in x] if isinstance(x, list) else x)
df['keywords'] = df['keywords'].apply(lambda x: [i.replace(" ","") for i in x] if isinstance(x, list) else x)
df['cast'] = df['cast'].apply(lambda x: [i.replace(" ","") for i in x] if isinstance(x, list) else x)
df['crew'] = df['crew'].apply(lambda x: [i.replace(" ","") for i in x] if isinstance(x, list) else x)
```

```python
df.head()
```

```python
df['tags'] = df['overview'].fillna('').apply(' '.join) + df['genres'].fillna('').apply(' '.join) + df['keywords'].fillna('').apply(' '.join) + df['cast'].fillna('').apply(' '.join) + df['crew'].fillna('').apply(' '.join)
```

```python
df.head()
```

```python
df_new = df[['movie_id','title','tags']]
```

```python
df_new
```

```python
df_new.info()
```

```python
df_new['tags'][0]
```

```python
df_new['tags'] = df_new['tags'].apply(lambda x: x.lower())
```

```python
df_new
```

# 3. Vectorization
The process of converting text data to numerical vectors

* Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(df_new['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

# 4. Stemming
The words “programming,” “programmer,” and “programs” can all be reduced down to the common word stem “program.

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
df_new['tags'] = df_new['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

# 5. Cosine Similarity
We consider Cosine Distance (theta between two vectors) and not Eucladian Distance, as it fails for higher dimensions because of 'Curse of Dimensionality'

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
# creating tuple of movie_id along with similarity count
# key=lambda x:x[1] specifies that sorting has to done on the basis of similarity not movie_id
```

```python
def recommend(movie):
    
    # to get movie index
    movie_index = df_new[df_new['title']== movie].index[0]
    
    distances = similarity[movie_index]
    
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(df_new.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
recommend('Iron Man')
```

```python
recommend('Life of Pi')
```
